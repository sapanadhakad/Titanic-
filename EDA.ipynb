```python
data['family_size'] = data['Parch']+data['SibSp']
data.sample(5)
```

```python
def family_type(number):
    if number == 0:
        return 'Alone'
    elif number > 0 and number <= 4:
        return 'Mediu'
    else:
        return 'Large'
```

```python
data['family_type'] = data[family_size].apply(family_type)
```

```python
data.sample(5)
```

```python
data.drop(columns = ['SibSp','Parch','family_size'],inplace = True)
```

```python
data.sample(5)
```
